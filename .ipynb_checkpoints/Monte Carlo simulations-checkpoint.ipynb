```python
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
from PIL import Image, ImageTk
import os

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
from scipy.stats import norm, logistic, norm
from scipy.special import ndtri, expit, erf, erfinv, erfc, erfcinv
import matplotlib.pyplot as plt
import datetime
import math
import time
import random
import glob

from basie_functions import *
```

```python
# MC setup
NMC=1000 # 1000 MC simulations
nt=300 # of 300 trials each

modelp = np.array([[0.5], [0.04], [0.2], [-20], [20], [0.01], [0.5]]) # thresh, miss, guess, minsnr, maxsnr, minslope, maxslope
basiep={'cs': 1}

nmodels=3

availsnr=np.linspace(-20, 20, 41).T
snrlist=[]
for i in range(nmodels):
    snrlist.append(availsnr)
snrlist=np.asarray(snrlist)

srterror=np.empty((NMC, nt,nmodels))
slopeerror=np.empty((NMC,nt,nmodels))
srtvar=np.empty((NMC,nt,nmodels))
slopevar=np.empty((NMC,nt,nmodels))

truemodel=np.array([[[0.75],[0.75],[0.75]], [[0.0], [4.0], [10.0]],
                    [[0.01],[0.01],[0.01]], [[0.02],[0.02],[0.02]],
                    [[0.2],[0.2],[0.2]], [[1],[1],[1]]]) # prob@thresh, thresh(dB), slope@thresh(db-1), prob(miss), prob(guess), fct type
```

```python
truemodel[:,1]
```

```python
%%capture
for imc in range(NMC):
    [snr, ii, __, __]=v_psycest(-nmodels,  np.repeat(modelp, nmodels, axis=1), basiep, snrlist,1)
    for i in range(nt):
        print(truemodel[:,ii-1])
        [response, __] = v_psychofunc('r',truemodel[:,ii-1],np.array([snr]));
        #response = np.array([[bool(listofresponses[i])]])
        # print('snr: ',snr, 'response: ',response, 'model: ', ii)
        [snr, ii, m, v] = v_psycest(ii, snr, response);
        srterror[imc, i] = m[0,:,0]-truemodel[1,:].flatten()
        slopeerror[imc, i] = m[1,:,0]-truemodel[2,:].flatten()
        srtvar[imc, i] = v[0,:]
        slopevar[imc, i] = v[2,:]
```

```python
ntests=300
response=np.empty((2,ntests))

for i in range(ntests):
    [response[:,i], __] = v_psychofunc('r',truemodel[:,2],np.array([[0],[10]]));
print(sum(response))
print(ntests-sum(response))
```

```python
srtRMSE=np.sqrt(np.mean(srterror**2, axis=0))
srtbias=np.mean(srterror, axis=0)
slopeRMSE=np.sqrt(np.mean(slopeerror**2, axis=0))
slopebias=np.mean(slopeerror, axis=0)

enssrtvar=np.mean(np.sqrt(srtvar), axis=0)
ensslopevar=np.mean(np.sqrt(slopevar), axis=0)

fig,axs=plt.subplots(2,3, constrained_layout=True)

axs[0][0].plot(range(nt),srtRMSE)
axs[0][0].set_xlabel('ntrials')
axs[0][0].set_ylabel('SRT RMSE')

axs[0][1].plot(srtbias)
axs[0][1].set_xlabel('ntrials')
axs[0][1].set_ylabel('SRT bias')

axs[0][2].plot(enssrtvar)
axs[0][2].set_xlabel('ntrials')
axs[0][2].set_ylabel('SRT ens. std')


axs[1][0].plot(slopeRMSE)
axs[1][0].set_xlabel('ntrials')
axs[1][0].set_ylabel('Slope RMSE')

axs[1][1].plot(slopebias)
axs[1][1].set_xlabel('ntrials')
axs[1][1].set_ylabel('Slope bias')

axs[1][2].plot(ensslopevar)
axs[1][2].set_xlabel('ntrials')
axs[1][2].set_ylabel('Slope ens. std')

plt.show()
```
